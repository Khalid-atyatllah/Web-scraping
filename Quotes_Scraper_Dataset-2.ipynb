```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd

data_list = []

def get_quotes(all_pages):
    page = 1
    while page <= all_pages:
        get_source = requests.get('http://quotes.toscrape.com/page/' + str(page) + '/')
        soap_scrape = BeautifulSoup(get_source.text, 'html.parser')
        
        items = soap_scrape.find_all('div', class_='quote')
        for article in items:
            author = article.find('small', class_="author").get_text()
            quote_elem = article.find('span', class_="text")
            quote = quote_elem.get_text() if quote_elem else "No quote found"
            
            tags_container = article.find('div', class_='tags')
            tags = tags_container.find_all('a', class_='tag')
            tag_text = [tag.get_text() for tag in tags]
            
            data = [author, ', '.join(tag_text), quote]
            data_list.append(data)
            
        page += 1

get_quotes(10)

header = ['Author', 'Tags', 'Quote']

with open('QuotesScraperDataset.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(data_list)

df = pd.read_csv('QuotesScraperDataset.csv')
df
```
