```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd

data_list = []
def map_star_rating(star_rating):
    rating_map = {
        'One': '1/5',
        'Two': '2/5',
        'Three': '3/5',
        'Four': '4/5',
        'Five': '5/5'
    }
    return rating_map.get(star_rating, 'Unknown')

def get_books(all_pages):
    base_url = 'http://books.toscrape.com/catalogue/category/books_1/page-{}.html'

    for page in range(1, all_pages + 1):
        url = base_url.format(page)
        get_source = requests.get(url)
        soap_scrape = BeautifulSoup(get_source.text, 'html.parser')

        books = soap_scrape.find_all('li', class_='col-xs-6 col-sm-4 col-md-3 col-lg-3')
        for book in books:
            title = book.h3.a['title']
            image_url = book.find('img')['src']
            star_rating = map_star_rating(book.find('p', class_='star-rating')['class'][1])
            price = book.select_one('div p.price_color').get_text()
            availability = book.select_one('div p.availability').get_text().strip()
            
            data = [title, image_url, star_rating, price, availability]
            data_list.append(data)

get_books(50)

header = ['Title', 'Image URL', 'Star Rating', 'Price', 'Availability']

with open('BooksScraperDataset.csv', 'w', newline='', encoding='UTF8') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(data_list)

df = pd.read_csv('BooksScraperDataset.csv')
df
```
